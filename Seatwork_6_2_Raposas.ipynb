# Seatwork 6.2 Programming Exercise: Getting Started with Pandas!

Using the 2019_Yellow_Taxi_Trip_Data.csv Links to an external site. dataset, accomplish the following items and submit a PDF of the notebook:

    1. Create a DataFrame by reading in the 2019_Yellow_Taxi_Trip_Data.csv file. Examine the first 5 rows.
    2. Find the dimensions (number of rows and number of columns) in the data.
    3. Using the data in the 2019_Yellow_Taxi_Trip_Data.csv file, calculate summary statistics for the fare_amount, tip_amount, tolls_amount, and total_amount columns.
    4. Isolate the fare_amount, tip_amount, tolls_amount, and total_amount for the longest trip by distance (trip_distance).


```python
import pandas as pd
from google.colab import files
```

```python
print("Upload your 2019_Yellow_Taxi_Trip_Data.csv file:")
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
```

```python
df = pd.read_csv(filename)
```

```python
print("\nFirst 5 rows of the dataset (selected columns):")
print(df[['fare_amount', 'tip_amount', 'tolls_amount', 'total_amount', 'trip_distance']].head().to_string(index=False))
```

```python
rows, columns = df.shape
print(f"\nDimensions of the dataset: {rows} rows, {columns} columns")
```

```python
summary = df[['fare_amount', 'tip_amount', 'tolls_amount', 'total_amount']].describe().round(2)

summary_display = summary.style.format("${:,.2f}")

print("\nSummary statistics for fare_amount, tip_amount, tolls_amount, total_amount:")
display(summary_display)
```

```python
longest_trip = df.loc[df['trip_distance'].idxmax(),
                      ['fare_amount', 'tip_amount', 'tolls_amount', 'total_amount', 'trip_distance']].round(2)

print("\nFare, tip, tolls, total amount, and distance for the longest trip:")
print(longest_trip.to_string())
```
