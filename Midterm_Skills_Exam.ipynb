```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import io

upload = widgets.FileUpload(
    accept='.csv,.json',
    multiple=True
)

display(upload)
```

```python
sales_data = None
customer_data = None

for filename, file_info in upload.value.items():

    if filename.endswith('.csv'):
        sales_data = pd.read_csv(io.BytesIO(file_info['content']))
        print("Sales data loaded")
        display(sales_data.head())

    elif filename.endswith('.json'):
        customer_data = pd.read_json(io.BytesIO(file_info['content']))
        print("Customer data loaded")
        display(customer_data.head())
```
