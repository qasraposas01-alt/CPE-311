```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

```python
print(salaries)
print(f"Number of Entries: {len(salaries)}")
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the
standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are
obtained when using the statistics module (where possible):

```python
#Calculating the Mean
mean = sum(salaries) / len(salaries)
print(f"{sum(salaries)} / {len(salaries)}")
print(f"The Mean: {mean}")
```

```python
#Calculating the Median
sorted_data = sorted(salaries)
print(f"Sorted Data:{sorted_data}")
mid1_ind = len(salaries) // 2 + 1
mid2_ind = len(salaries) // 2
median = (mid1_ind + mid2_ind) // 2
print(f"Middle values:{salaries[mid1_ind]},{salaries[mid2_ind]}")
print(f"Median of Data:{median}")
```

```python
#Getting the Mode
map_val = {}
for salary in salaries:
  if salary not in map_val:
    map_val[salary] = 1
    continue
  map_val[salary] += map_val[salary] + 1

map_val
mode = max(map_val, key=map_val.get)
print(f"Mode: {mode}")
print(f"Frequency: { max(map_val.values())}")
```

```python
#Sample variance
varval = []
for element in salaries:
  varval.append(((element - mean)**2) /(len(salaries)-1))
variance = sum(varval)
print(f"Sample Variance: {variance}")

#Sample Standard Deviation
standardDev = variance ** 0.5
print(f"Sample Standard Deviation: {standardDev}")
```

##Excercise 3: Pandas for Data Analysis

```python
#Load the diabetes.csv file. Convert the diabetes.csv into dataframe
import pandas as pd
import numpy as np
data = pd.read_csv('diabetes.csv')
df = pd.DataFrame(data)
df.head()
```

```python
# Identify the column names
print(f"Dataset Column Names: {list(df.columns)}")
print("-----" * 20)
# Identify the data types of the data
print(f"\n Datatypes per column: \n{(df.dtypes)}")
print("-----" * 20)
# Display the total number of records
print(f"\n Total Records: {len(df)}")
print("-----" * 20)
```

```python
# Display the first 20 records
df.head(20)
```

```python
# Display the last 20 records
df.tail(20)
```

```python
# Change the Outcome column to DiagnosisIn [ ]:
renamed_df = df.rename(columns = {'Outcome' : 'DiagnosisIN'})
print(f"New Dataset Column Names: {list(renamed_df.columns)}")
```

```python
#Create a new column Classification that display "Diabetes" if the value of outcome is 1 , otherwise "No Diabetes"
df['Classification'] = np.where(df['Outcome'] == 1, 'Diabetes', 'No Diabetes')
df.head()
```

```python
# Create a new dataframe "withDiabetes" that gathers data with diabetes
withDiabetes = df[df['Classification'] == 'Diabetes']
withDiabetes.head()
```

```python
# Create a new dataframe "noDiabetes" thats gathers data with no diabetes
noDiabetes = df[df['Classification'] == 'No Diabetes']
noDiabetes.head()
```

```python
# Create a new dataframe "Pedia" that gathers data with age 0 to 19
pedia = df[(df['Age'] > 0) & (df['Age'] < 19)]
pedia.head()
```

```python
# Create a new dataframe "Adult" that gathers data with age greater than 19
adult = df[df['Age'] > 19]
adult.head()
```

```python
# Use numpy to get the average age and glucose value.
aveAge = np.mean(df['Age'])
print(f'Average of Age: {aveAge}')
aveGlucose = np.mean(df['Glucose'])
print(f'Average of Glucose: {aveGlucose}')
```

```python
# Use numpy to get the median age and glucose value.
medianAge = np.median(df['Age'])
print(f'Median of Age: {medianAge}')
medianGlucose = np.median(df['Glucose'])
print(f'Median of Glucose: {medianAge}')
```

```python
# Use numpy to get the middle values of glucose and age.
medGlucose = np.median(df['Glucose'])
print(f'Median of Glucose: {medGlucose}')
medAge = np.median(df['Age'])
print(f'Median of Age: {medAge}')
```

```python
# Use numpy to get the standard deviation of the skinthickness.
stdST = np.std(df['SkinThickness'])
print(f'Standard Deviation of Skin Thickness: {stdST}')
```
